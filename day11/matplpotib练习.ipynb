```python
import matplotlib.pyplot as plt
```

# 折线图

```python
# 单折线图
import random
# x从2到26的范围内取，步长为2结果为2-24
x = range(2,26,2)
# 为每个x分配一个15-30之内的随机数
y = [random.randint(15,30) for i in x]
# 将x轴的刻度转变为i:00并让其旋转45°
x_ticks_label = [f"{i}:00" for i in x]
plt.xticks(x,x_ticks_label,rotation = 45)
# y轴的刻度  为每个刻度增加℃，并且因为range的结尾是开区间，+1使得max(y)能被包括到图表里
y_ticks_label = ["{}℃".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_ticks_label)
plt.plot(x,y)
plt.show()
```

```python
# 显示中文
x=range(0,120)
# 为y分配120个在10-30内的随机数
y=[random.randint(10,30) for i in range(120)]
# 配置画布
plt.figure(figsize=(20,8),dpi=80)
# 导入字体
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname=r'C:\Windows\Fonts\simsun.ttc', size=20)
# 为 x 轴和 y 轴分别设置标签
plt.xlabel('时间',rotation=45,fontproperties=my_font)
plt.ylabel("次数",fontproperties=my_font)
# 设置标题
plt.title('每分钟跳动次数',color='blue',fontproperties=my_font)
plt.plot(x,y)
```

# 散点图

```python
y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17,
     18, 21, 16, 17, 20, 14, 15, 15, 15, 19, 21, 22, 22, 22, 23]
x =range(1,32)
plt.figure(figsize=(20,8),dpi=80)
# 使用scatter绘制散点图
plt.scatter(x,y,label='2月份')
my_font = font_manager.FontProperties(fname=r'C:\Windows\Fonts\simsun.ttc', size=20)
# 调整x轴刻度
x_ticks_label=['2月{}日'.format(i)for i in x]
# 对x轴进行每三步一次的切片
plt.xticks( x[::3], x_ticks_label[::3], fontproperties=my_font, rotation=45 )
plt.xlabel( '日期', fontproperties=my_font )
plt.ylabel( '温度', fontproperties=my_font )
plt.grid( alpha=0.4 )
plt.legend( prop=my_font )
plt.show()
```

# 条形图

```python
my_font =  font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc',size=16)
a = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
b = [38.13,19.85,14.89,11.36,6.47,5.93]
plt.figure(figsize=(20,8),dpi=80) 
# 绘制条形图
# x为0-len(a)-1   y为b  宽度为0.3
rects = plt.bar(range(len(a)),b,width=0.3,color='r')
# 设置x轴的标签
plt.xticks(range(len(a)),a,fontproperties=my_font)
for rect in rects:
    height = rect.get_height() # 获取条形的高度
    # rect.get_x() + rect.get_width() / 2: 条形的 x 坐标加上条形宽度的一半，使文本居中。
    # height + 0.5: 条形顶部高度加上 0.5，使文本稍微高于条形顶部。
    # '%.2f' % height: 将高度格式化为小数点后两位。
    # ha='center': 设置水平对齐为居中。
    # va='bottom': 设置垂直对齐为底部。
    plt.text(rect.get_x()+rect.get_width()/2,height+0.5,'%.2f'%height,
             ha='center',va='bottom',fontproperties=my_font)
plt.show()
```

# 饼图

```python
 # 各部分标签
label_list = ["第一部分", "第二部分", "第三部分"]
# 各部分大小
size = [55, 35, 10]
# 各部分颜色
color = ["red", "green", "blue"]
# 各部分突出值
explode = [0, 0, 0]
plt.figure(figsize=(20, 8), dpi=100)
# 设置字体
plt.rcParams['font.sans-serif']=['SimHei']
# 绘制饼图
patches, l_text, p_text = plt.pie(size,explode=explode, colors=color, labels=label_list,
                                  labeldistance=1.1, autopct="%1.1f%%", shadow=True, startangle=90, pctdistance=0.5)
# 设置横轴和纵轴大小相等，这样饼才是圆的
plt.axis("equal")
plt.legend()
plt.show()
```
